```python
#
# 远亲不如近邻
# @param n int整型 居民个数
# @param m int整型 方案个数
# @param a int整型一维数组 居民的位置
# @param x int整型一维数组 方案对应的位置
# @return int整型一维数组
#


class Solution:
    # 二分法
    def nearest(self, n, a, x):
        d = pow(2, 31)
        distance = d
        l = 0
        r = n
        index = int((l+r)/2)
        if n == 1:
            return abs(a[0]-x)
        if n == 2:
            return min(abs(a[0]-x), abs(a[1]-x))
        
                       
        while r-l>1:
            index=int((l+r)/2)
#             print("l=", l, "r=", r, "index=", index)
            d=a[index]-x
            
            if d >= 0:
                r=index
            if d < 0:
                l=index
        
        if d>=0:
            distance=min(d,abs(a[max(index-1,0)]-x))
        else:
            distance=min(-d,abs(a[min(n-1,index+1)]-x))

        return distance

    def solve(self, n, m, a, x):
        # write code here
        a.sort()
        ans=[-1 for i in range(m)]
        for i in range(m):
            ans[i]=self.nearest(n=n, a=a, x=x[i])

        return ans
```
